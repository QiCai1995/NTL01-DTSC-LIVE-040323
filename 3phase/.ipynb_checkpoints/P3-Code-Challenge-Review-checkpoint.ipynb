### 3) What is a step size when talking about gradient descent? How does learning rate regulate step size?

```python
# call_on_students(1)

# We always want to find the minimum loss in a model, gradient decent is trying to find the minimum cost or loss.
# The size of the steps (α, alpha) is proportional to the value of the derivative.
# The size of the steps or alpha is the learning rate.
# If alpha is higher big step. If alpha is lower small step.
# New step is old step minus derivative of the old step times the learning rate.


# The step size is also known as the learning rate in terms of gradient descent.
#  The higher the learning rate, the larger step sizes, causing the algorithm to make more significant updates to
# the parameters at each iteration. The lower learning rate leads to smaller step sizes and slower convergence.

# SOLUTION
# Step size is the partial derivative of the curve
# learning rate is the amount of the step size to actually take.
```

## Part 3: Logistic Regression and Classification Metrics with Code

### Set Up

```python
# Let's use the same data, but now with a classification target
mpg_class = pd.read_csv('data/auto-mpg-classification.csv')
```

```python
# Check this new dataframe out
mpg_class.head()
```

### 6) Prepare our data for modeling:

1. Perform a train/test split
2. Scale the inputs


```python
# Train-test split
# Set test_size=0.33 and random_state=42

X = mpg_class.drop(columns = 'target')
y = mpg_class['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

```python
# Scale the data
ss = StandardScaler()
X_train_transformed = ss.fit_transform(X_train)
X_test_transformed = ss.transform(X_test)

```

### 9) Fit a logistic regression model, and plot a confusion matrix of the results on our test set

```python
# call_on_students(1)
```

```python
# Fit a logistic regression model
# Name the model `logreg` and set random_state = 42
logreg = LogisticRegression(random_state = 42)
logreg.fit(X_train_transformed, y_train)
y_test_pred = logreg.predict(X_test_transformed)
```

```python
# Plot a confusion matrix on the test data
plot_confusion_matrix(logreg, X_test_transformed, y_test)
```

### 10) Calculate the accuracy, precision, recall and f1-score for the test set

You can use the confusion matrix above, or sklearn functions

```python
# call_on_students(1)
```

```python
# Accuracy
accuracy_score(y_test, y_test_pred)
```

```python
# Precision
precision_score(y_test, y_test_pred)
```

```python
# Recall
recall_score(y_test, y_test_pred)
```

```python
# F1-Score
f1_score(y_test, y_test_pred)
```

### 11) Calculate the ROC-AUC on the test set, and plot the ROC curve

For this you'll definitely want to use the sklearn functions!

```python
# call_on_students(1)
```

```python
# Calculate roc-auc
y_probs = logreg.predict_proba(X_test_transformed)
roc_auc_score(y_test, y_probs[:,1])
```

```python
# Plot the ROC curve
plot_roc_curve(logreg, X_test_transformed, y_test)
```

## Part 4: Decision Trees

### Set Up - Generating 1-split trees to compare

```python
# Creating two different decision trees with a single split
dt_maxdepth1_v1 = DecisionTreeClassifier(max_depth=1, random_state = 42)
dt_maxdepth1_v2 = DecisionTreeClassifier(max_depth=1, random_state = 42)

# Training the two trees on different columns
dt_maxdepth1_v1.fit(X_train[['weight']], y_train)
dt_maxdepth1_v2.fit(X_train[['origin']], y_train)
```

```python
# Tree based on 1st column
plot_tree(dt_maxdepth1_v1, feature_names=['weight'], filled=True);
```

```python
# Tree based on 2nd column
plot_tree(dt_maxdepth1_v2, feature_names=['origin'], filled=True);
```

### 14) Fit a decision tree model, and plot a confusion matrix of the results on our test set

```python
# call_on_students(1)
```

```python
# Fit a decision tree model
# Name the model `dt` and set random_state = 42

dt = DecisionTreeClassifier(random_state = 42)

dt.fit(X_train_transformed, y_train)
```

```python
# Plot a confusion matrix on the test data
plot_confusion_matrix(dt, X_test_transformed, y_test)
plt.show()
```

```python
# Code to visualize the ROC curves for the models we've done
fig,ax = plt.subplots()
plot_roc_curve(dt, X_test_transformed, y_test, ax = ax)
plot_roc_curve(logreg, X_test_transformed, y_test, ax = ax)

plt.title("Receiver Operating Characteristic Curves\n(Evaluated on Test Set)")
plt.show()

```
